```python
import struct

PAGE_SIZE = 16   # bigger page: 4 bytes number + 12 bytes name

with open("test_page.db", "wb") as f:       # open ONCE outside loop
    for page_no in range(5):
        page = bytearray(PAGE_SIZE)          # fresh blank page

        # write page number at bytes 0-3
        page[0:4] = struct.pack("I", page_no)

        # write page name at bytes 4-10
        name = f"page_{page_no}".encode("utf-8")  # 6 bytes
        page[4: 4 + len(name)] = name

        f.write(page)                        # append to file, no wiping!
```

```python
with open("test_page.db", "rb") as f:
    for page_no in range(5):
        page = f.read(PAGE_SIZE)
        num  = struct.unpack("I", page[0:4])[0]
        name = page[4:10].decode("utf-8").rstrip("\x00")
        print(f"Page {num} → {name}")

# Output:
# Page 0 → page_0
# Page 1 → page_1
# Page 2 → page_2
# Page 3 → page_3
# Page 4 → page_4
```
